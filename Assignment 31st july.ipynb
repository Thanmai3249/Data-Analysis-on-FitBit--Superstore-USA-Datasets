```python
returns = pd.read_csv('C:/Users/Thanmai/INEURON//returns.csv')
returns = returns.drop(['Unnamed: 0'], axis = 1)

import mysql.connector as connection
from mysql.connector import Error

mydb = connection.connect(host='localhost', database='Superstore', user='root', password='thanmai123')
if mydb.is_connected():
    cursor = mydb.cursor()
    cursor.execute("select database();")
    record = cursor.fetchone()
    print("You're connected to database: ", record)
    cursor.execute('DROP TABLE IF EXISTS returns;')
    print('Creating table....')
# in the below line please pass the create table statement which you want #to create
cursor.execute('''create table returns(
                    OrderID       varchar(50),
                    Status        varchar(500)
                    );
                    ''')
print("Table is created....")
```

```python
import mysql.connector as connection
mydb = connection.connect(host='localhost', database='Superstore', user='root', password='thanmai123')
q3 = '''
SET GLOBAL FOREIGN_KEY_CHECKS=0;
SET FOREIGN_KEY_CHECKS=0;
ALTER TABLE `Superstore`.`orders` 
ADD INDEX `Order ID_idx` (`Order ID` ASC) VISIBLE;
;
ALTER TABLE `Superstore`.`orders` 
ADD CONSTRAINT `Order ID`
  FOREIGN KEY (`Order ID`)
  REFERENCES `Superstore`.`returns` (`Order ID`)
  ON DELETE NO ACTION
  ON UPDATE NO ACTION;
  ALTER TABLE `Superstore`.`orders` 
ADD INDEX `Region_idx` (`Region` ASC) VISIBLE;
;
ALTER TABLE `Superstore`.`orders` 
ADD CONSTRAINT `Region`
  FOREIGN KEY (`Region`)
  REFERENCES `Superstore`.`users` (`Region`)
  ON DELETE NO ACTION
  ON UPDATE NO ACTION;
'''

cursor = mydb.cursor() 
cursor.execute(q3)
print(cursor.fetchall())
```
